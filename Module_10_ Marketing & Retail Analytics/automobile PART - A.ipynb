```python
# Importing the necessary libraries
import pandas as pd            # For data manipulation
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns          # For advanced visualizations
```

```python
# Specify the file path
file_path = r"C:\Users\GNANASEKARAN\OneDrive\Documents\Python Project\Sales_Data.xlsx"

# Load the data from the Excel file
data = pd.read_excel(file_path)
```

```python
# Display the first few rows of the data
print(data.head())
```

```python
# Display the first few rows of the data
print(data.tail())
```

```python
# Check the shape of the data (number of rows and columns)
print(data.shape)
```

```python
# Get the data description 
print(data.info())
```

```python
# Check for missing values in each column
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Check for duplicate rows
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
# Get the statistical summary of the data
summary = data.describe()
print(summary)
```

```python
#Univariate analysis
# Plotting histograms for numerical columns
numerical_columns = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP', 'DAYS_SINCE_LASTORDER']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[col], bins=20, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plotting bar charts for categorical columns
categorical_columns = ['STATUS', 'PRODUCTLINE', 'COUNTRY']

plt.figure(figsize=(12, 8))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=data[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    
    # Rotate X-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
#Bivariate analysis
# Scatter plot between 'SALES' and 'QUANTITYORDERED'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='QUANTITYORDERED', y='SALES', data=data)
plt.title('Sales vs. Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.show()

# Correlation heatmap of numerical variables
corr_matrix = data[numerical_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Boxplot for SALES vs PRODUCTLINE
plt.figure(figsize=(10, 6))
sns.boxplot(x='PRODUCTLINE', y='SALES', data=data)
plt.title('Sales vs Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#multivariate analysis
# Pairplot of numerical columns to visualize interactions
sns.pairplot(data[numerical_columns])
plt.show()
```

```python
# FacetGrid for 'SALES' by 'PRODUCTLINE' and 'STATUS'
g = sns.FacetGrid(data, col="PRODUCTLINE", row="STATUS", margin_titles=True)
g.map(sns.scatterplot, "QUANTITYORDERED", "SALES")
g.set_axis_labels("Quantity Ordered", "Sales")
plt.show()
```

```python
#Trends Analysis (Weekly, Monthly, Quarterly, Yearly Trends in Sales)
# Convert ORDERDATE to DateTime
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
# Create a new column for week
data['Week'] = data['ORDERDATE'].dt.isocalendar().week

# Weekly trend of sales
weekly_sales = data.groupby('Week')['SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Week', y='SALES', data=weekly_sales)
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Create a new column for month
data['Month'] = data['ORDERDATE'].dt.month

# Monthly trend of sales
monthly_sales = data.groupby('Month')['SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='SALES', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Create a new column for quarter
data['Quarter'] = data['ORDERDATE'].dt.quarter

# Quarterly trend of sales
quarterly_sales = data.groupby('Quarter')['SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Quarter', y='SALES', data=quarterly_sales)
plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Create a new column for year
data['Year'] = data['ORDERDATE'].dt.year

# Yearly trend of sales
yearly_sales = data.groupby('Year')['SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='SALES', data=yearly_sales)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```
